```julia
using DataFrames, CSV
class_data = CSV.read("fireroading_updated_prereq_v5.csv", DataFrame)
```

```julia
# Enter your year, major and classes taken so far:

class_year = 2 #1 = Freshman, 2 = Sophomore, 3 = Junior (Assuming Spring semester start)

major_minors = 0 #by 0 = 6-14; 1 = 2 + Japanese]

classes_taken = [(45,1),(51,1),(49, 1),(8,1),(15,2),(7,2), (50,2),(52, 2)] #provided in tuple form:(subject number, semester taken with 1 being freshman fall)

#Student Preferences:

max_hours = [50,50,30,40,40,40,40,40] #Provided in the form where the ith index denotes the max number of hours in the ith semester
min_hours = [5,5,5,5,5,5,5,5]

#Bad Timeslots:
bad_timeslots = [0 for i in 1:80] #Students place a 1 in any timeslot that they do not want classes in for the upcoming semester

#Requirements for classes taken in a specific semester
class_requests = [] #provided in tuple form:(subject number, semester taken with 1 being freshman fall)
```

```julia
#--- Model specification
using JuMP, DataFrames, Gurobi

fireroading_optimizer = Model(with_optimizer(Gurobi.Optimizer))


@variable(fireroading_optimizer, x[i = 1:54, j = 1:8], binary = true)

#Make sure that only future classes are changed and all previous classes are accounted for:
for semester in 1:class_year * 2 - 1
    for class in 1:54
        if (class,semester) in classes_taken
            @constraint(fireroading_optimizer, x[class, semester] == 1) 
        else
            @constraint(fireroading_optimizer, x[class, semester] == 0) 
        end
    end
end



#HASS constraint
@constraint(fireroading_optimizer, sum(x[i, j] * class_data[!, :"is_hass"][i] for i in 1:54 for j in 1:8) >= 8)


#GIR constraint
@constraint(fireroading_optimizer, sum(x[i,j] * class_data[!, :"is_gir"][i] for i in 1:54 for j in 1:8) >= 6)

#Major Constraints:

#6-14 Requirements:
if major_minors == 0
    @constraint(fireroading_optimizer, sum(x[21,j] for j in 1:8) >= 1) #18.06 requirement
    @constraint(fireroading_optimizer, sum(x[8,j] for j in 1:8) >= 1) #6.00 requirement
    @constraint(fireroading_optimizer, sum(x[24,j] for j in 1:8) >= 1) #6.006 requirement
    @constraint(fireroading_optimizer, sum(x[7,j] for j in 1:8) >= 1) #6.042 requirement
    @constraint(fireroading_optimizer, sum(x[1,j] for j in 1:8) >= 1) #6.046 requirement
    @constraint(fireroading_optimizer, sum(x[15,j] for j in 1:8) >= 1) #6.009 requirement
    @constraint(fireroading_optimizer, sum(x[45,j] for j in 1:8) >= 1) #14.01 requirement
    @constraint(fireroading_optimizer, sum(x[27,j] for j in 1:8) >= 1) #14.32 requirement
    @constraint(fireroading_optimizer, sum(x[37,j] for j in 1:8) >= 1) #14.30 requirement
    @constraint(fireroading_optimizer, sum(x[22,j] for j in 1:8) >= 1) #6.036 requirement
    @constraint(fireroading_optimizer, sum(x[26,j] for j in 1:8) >= 1) #6UAR requirement
    @constraint(fireroading_optimizer, sum(x[9,j] + x[16,j] for j in 1:8) >= 1) #14.05 or 14.18 requirement
    @constraint(fireroading_optimizer, sum(x[25,j] + x[46,j] for j in 1:8) >= 1) #6.207 or 6.215 requirement
    @constraint(fireroading_optimizer, sum(x[25,j] + x[46,j] for j in 1:8) >= 1) #6.207 or 6.215 requirement
    @constraint(fireroading_optimizer, sum(x[43,j] + x[11,j] + x[47,j] for j in 1:8) >= 1) #14.64 or 14.75 or 14.41 requirement
    @constraint(fireroading_optimizer, sum(x[42,j] + x[18,j] + x[34,j] for j in 1:8) >= 1) #14.19 or 14.26 or 14.54 requirement
    @constraint(fireroading_optimizer, sum(x[43,j] + x[11,j] + x[47,j] + x[42,j] + x[18,j] + x[34,j] for j in 1:8) >= 3) #3 or more of: 14.64 or 14.75 or 14.41 or 14.19 or 14.26 or 14.54 requirement

end

#Course 2 Major and Japanese Minor Requirements:
if major_minors == 1
    #Course 2 Requirements
    @constraint(fireroading_optimizer, sum(x[5,j] for j in 1:8) >= 1) # 2.001 requirement
    @constraint(fireroading_optimizer, sum(x[48,j] for j in 1:8) >= 1) # 2.002 requirement
    @constraint(fireroading_optimizer, sum(x[28,j] for j in 1:8) >= 1) # 2.003 requirement
    @constraint(fireroading_optimizer, sum(x[14,j] for j in 1:8) >= 1) # 2.004 requirement
    @constraint(fireroading_optimizer, sum(x[32,j] for j in 1:8) >= 1) # 2.005 requirement
    @constraint(fireroading_optimizer, sum(x[44,j] for j in 1:8) >= 1) # 2.006 requirement
    @constraint(fireroading_optimizer, sum(x[20,j] + x[2,j] for j in 1:8) >= 1) # 2.007/2.017 requirement
    @constraint(fireroading_optimizer, sum(x[3,j] for j in 1:8) >= 1) # 2.008 requirement
    @constraint(fireroading_optimizer, sum(x[30,j] for j in 1:8) >= 1) # 2.086 requirement
    @constraint(fireroading_optimizer, sum(x[19,j] for j in 1:8) >= 1) # 2.670 requirement
    @constraint(fireroading_optimizer, sum(x[23,j] for j in 1:8) >= 1) # 2.671 requirement
    @constraint(fireroading_optimizer, sum(x[17,j] for j in 1:8) >= 1) # 2.THU requirement
    @constraint(fireroading_optimizer, sum(x[38,j] for j in 1:8) >= 1) # 18.03 requirement
    @constraint(fireroading_optimizer, sum(x[29,j] + x[36,j] for j in 1:8) >= 1) # 2.009 or 2.013 requirement
    @constraint(fireroading_optimizer, sum(x[35,j] + x[31,j] + x[4,j] for j in 1:8) >= 2) # 2.014, 2.016, 2.019 requirement

    #Begin Japanese Minor:
    @constraint(fireroading_optimizer, sum(x[12,j] for j in 1:8) >= 1) # 21G.503
    @constraint(fireroading_optimizer, sum(x[13,j] for j in 1:8) >= 1) # 21G.504 requirement
    @constraint(fireroading_optimizer, sum(x[40,j] for j in 1:8) >= 1) # 21G.505 requirement
    @constraint(fireroading_optimizer, sum(x[10,j] for j in 1:8) >= 1) # 21G.506 requirement
    @constraint(fireroading_optimizer, sum(x[33,j] + x[6,j] + x[41,j] + x[39,j] for j in 1:8) >= 2) # 17.433, 21H.354, 21G.591, 21G.593 requirement
    @constraint(fireroading_optimizer, sum(x[41,j] + x[39,j] for j in 1:8) >= 1) # 21G.591, 21G.593 requirement

end    

# Prerequisite requirement:
for i in 1:54
    for j in 1:8
        #Require students to only take classes that have prereqs satisfied
        if sum(class_to_prereq[class_data[!, :"subject_id"][i]])>0
            if j > 1
                @constraint(fireroading_optimizer, .9 * x[i,j]  <=  sum(class_to_prereq[class_data[!, :"subject_id"][i]][l] * x[l,m] for l in 1:54 for m in 1:j-1)/sum(class_to_prereq[class_data[!, :"subject_id"][i]]))
            else
                @constraint(fireroading_optimizer, x[i,j]  <= 0 )    
            end
        end
    end
end

#216 Units beyond GIRs:
@constraint(fireroading_optimizer, sum(x[i,j]*(1-class_data[!, :"is_gir"][i])* class_data[!, :"units"][i] for i in 1:54 for j in 1:8) >= 216)
    

# #Time overlap constraint:
for t in 1:80
    @constraint(fireroading_optimizer, sum(x[i, class_year*2] * class_to_schedule[class_data[!, :"subject_id"][i]][t] for i in 1:54)<= 1)
end


#Each class can be taken at most once:
for i in 1:54
    @constraint(fireroading_optimizer, sum(x[i, j] for j in 1:8) <= 1)
end

#Student Preferences:

#Hours constraint:
for j in 1:8
    @constraint(fireroading_optimizer, sum(x[i, j]*class_data[!, :"hours"][i] for i in 1:54) <= max_hours[j])
end

#Bad Timeslot constraint:
for t in 1:80
    @constraint(fireroading_optimizer, sum(x[i, class_year*2] * bad_timeslots[t] for i in 1:54)<= 0)
end

#Requirement for subjects taken in specific semester
for class in class_requests
    @constraint(fireroading_optimizer, x[class[1], class[2]] == 1)
end




```
